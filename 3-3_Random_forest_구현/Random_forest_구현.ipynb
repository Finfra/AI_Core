```
from sklearn.ensemble import RandomForestClassifier
from sklearn import datasets
iris = datasets.load_iris()

X = iris.data[:, :4]
y = iris.target
model = RandomForestClassifier(n_estimators=15, max_depth=10, random_state=5)
model.fit(X, y)
```

# Accuracy

```
from sklearn.metrics.cluster import completeness_score
y_predict = model.predict(X[:, : ])
print("%s %s"%("Train Accuracy" , completeness_score(y, y_predict ) ))
```

# Confusion Matrix

```
import pandas as pd
pd.crosstab(y,y_predict)
```

---
## Over fitting이 의심됨 
→ traing data와 test data 분리 해서 진행


```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 2)

model = RandomForestClassifier(n_estimators=15, max_depth=10, random_state=5)
model.fit(x_train, y_train)


```

# Accuracy

```
from sklearn.metrics.cluster import completeness_score
y_predict2 = model.predict(  x_test)
print("%s %s"%("Test Accuracy" , completeness_score(y_test, y_predict2 ) ))
```

# Confusion Matrix

```
import pandas as pd
pd.crosstab(y_test,y_predict2)
```
