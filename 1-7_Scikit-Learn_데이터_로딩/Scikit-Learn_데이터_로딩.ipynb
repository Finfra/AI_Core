# CF : kNN
* Source : [scikit-learn을 활용한 기계 학습](https://github.com/brenden17/blog/blob/master/post/ms.scikit-learn.v.md)

```
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X, y)
sample = [[6, 4, 6, 2],]
predicted_value = knn.predict(sample)
print(iris.target_names[predicted_value])
print(knn.predict_proba(sample))
```
